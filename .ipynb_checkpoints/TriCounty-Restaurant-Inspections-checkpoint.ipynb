```python
# Dependencies
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

# Tri-County Restaurant Inspection Data

```python
# Read CSV
restaurants = pd.read_csv("Restaurant_Inspections_in_Tri-County_Colorado.csv")
```

```python
# Clean data
df = restaurants.fillna("0")
df.head()
```

```python
# Remove Agency column - all inspected by same agency. not useful information.
del df['Agency']
df.head()
```

```python
# Total Facility IDs
total_facilities = df["Facility ID"].nunique()
total_facilities
```

```python
# Remove Duplicates for most recent data
df_clean = df.drop_duplicates('Facility ID', keep='last')
df_clean.head()
```

```python
# Form new columns: totalInfractions & inspScore
df_clean['Total Violations'] = df_clean['Total Foodborne Illness Risk Violations'] + df_clean['Total Good Retail Practices Violations']
df_clean.loc[:,'inspScore'] = 56-df_clean['Total Violations']
df_clean.head()
```

```python
# Categorical Breakdown
PE_inspections = df_clean['PE Desc'].value_counts()
PE_inspections
```

```python
# Group by and filter by Restaurant size, Risk factor
restaurant_risk = df_clean.groupby(["PE Desc","Category"])['Program Identifier'].count()
restaurant_risk
```

```python
# Group by county, city
geographic_df = df_clean.groupby(["County", "City"])['Program Identifier'].count()
geographic_df
```
